```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
sns.set()
```

Data Collection

```python
books_data=pd.read_csv("/content/bestsellers with categories.csv")
```

```python
#print first 5 rows
books_data.head()
```

```python
books_data.isnull().sum()
```

```python
books_data["Name"].value_counts()
```

```python
books_data.shape
```

```python
books_data.columns
```

```python
plot=plt.figure(figsize=(15,5))
plt.bar(list(books_data["Name"].value_counts()[:3].keys()),list(books_data["Name"].value_counts()[:3]))

plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.set_style(style="darkgrid")
sns.countplot(y="User Rating",data=books_data,palette="Set2", order=books_data["User Rating"].value_counts().index[:15])
```

```python
books_data["Genre"].value_counts()
```

```python
plt.figure(figsize=(10,5))
sns.countplot("Genre",data=books_data)
```

```python
#check the price
books_data["Price"].value_counts()
```

```python
plt.figure(figsize=(10,7))
plt.hist(books_data["Price"])
```

```python
#print top 10 costly books price
plt.figure(figsize=(10,6))
sns.countplot(x="Price", data=books_data, palette="Set2", order=books_data["Price"].value_counts().index[:10])
```

```python
#check author count
books_data["Author"].value_counts()
```

```python
#draw barplot for top 5 authors
plt.figure(figsize=(15,5))
plt.bar(list(books_data["Author"].value_counts()[:5].keys()),list(books_data["Author"].value_counts()[:5]))
```

```python
#draw piechart 
plt.figure(figsize=(10,8))
plt.pie(list(books_data["Genre"].value_counts()),labels=list(books_data["Genre"].value_counts().keys()),autopct="%.1f%%")
plt.show()
```

```python
#name
books_data["Name"].value_counts()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x="Price",hue="Genre",palette="rainbow", data=books_data)
plt.show()
```

```python
sns.distplot(books_data["Year"],kde=False)
plt.show()
```
